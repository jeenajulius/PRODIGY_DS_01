```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('/content/Infant_Mortality_20231205.csv')
```

```python
grouped_data = df.groupby('Materal Race or Ethnicity')['Infant Mortality Rate'].mean().sort_values()
```

```python
plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar', color='skyblue')
plt.title('Average Infant Mortality Rate by Race or Ethnicity')
plt.xlabel('Race or Ethnicity')
plt.ylabel('Average Infant Mortality Rate')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
