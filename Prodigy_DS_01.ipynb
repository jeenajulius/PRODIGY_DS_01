```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the paths to the extracted files
metadata_indicator_path = 'Metadata_Indicator_API_SP.POP.TOTL_DS2_en_csv_v2_2544372.csv'
data_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_2544372.csv'
metadata_country_path = 'Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_2544372.csv'

# Load each file into a pandas DataFrame
metadata_indicator = pd.read_csv(metadata_indicator_path)
data = pd.read_csv(data_path, skiprows=4)
metadata_country = pd.read_csv(metadata_country_path)

# Display the first few rows of each DataFrame to confirm loading
print("Metadata Indicator:")
print(metadata_indicator.head())

print("\nMain Data:")
print(data.head())

print("\nMetadata Country:")
print(metadata_country.head())

data_2023 = data[['Country Name', '2023']].dropna()
data_2023.columns = ['Country Name', 'Population_2023']

# Plot histogram of population sizes in 2023
plt.figure(figsize=(12, 6))
plt.hist(data_2023['Population_2023'] / 1e6, bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Population Sizes in 2023')
plt.xlabel('Population (in millions)')
plt.ylabel('Number of Countries')
plt.grid(True)
plt.show()
```
